```python
import pandas as pd

# Paths to the downloaded files
data_path = os.path.join(root_dir, "bank-additional-full.csv")

# Load data
df_initial = pd.read_csv(data_path, sep=';')

# Paths to the downloaded files
data_path = os.path.join(root_dir, "test.csv")

# Load data
df_test = pd.read_csv(data_path)
```

```python
df_test = df_test[['age', 'job', 'marital', 'education', 'default', 'housing',
       'loan', 'contact', 'month', 'day_of_week', 'campaign', 'pdays',
       'previous', 'poutcome', 'emp.var.rate', 'cons.price.idx',
       'cons.conf.idx', 'euribor3m', 'nr.employed', 'id']]
```

```python
df_initial = df_initial[['age', 'job', 'marital', 'education', 'default', 'housing',
       'loan', 'contact', 'month', 'day_of_week', 'campaign', 'pdays',
       'previous', 'poutcome', 'emp.var.rate', 'cons.price.idx',
       'cons.conf.idx', 'euribor3m', 'nr.employed', 'y']]
```

```python
df_initial.info()
```

```python
df_test['key'] = df_test.iloc[:, 0:19].astype(str).apply(''.join, axis=1)
df_initial['key'] = df_initial.iloc[:, 0:19].astype(str).apply(''.join, axis=1)
```

```python
merged_df = pd.merge(df_test, df_initial, on=['key', 'age', 'job', 'marital', 'education', 'default', 'housing',
       'loan', 'contact', 'month', 'day_of_week', 'campaign', 'pdays',
       'previous', 'poutcome', 'emp.var.rate', 'cons.price.idx',
       'cons.conf.idx', 'euribor3m', 'nr.employed'], how='inner')
```

```python
merged_df.info()
```

```python
merged_df = merged_df.drop_duplicates()
```
